```
# Номер задания 173
from tkinter import *
import random
window = Tk()
window.title("Прямоугольник")
can = Canvas(window, width = 500, height = 500, bg = 'white')
can.pack()
n = int(input("Введите натуральное число n: "))
arr = [random.randint(0, 500) for i in range(n)]
i = 0
while i < len(arr):
    if (i % 4 == 0) and (i+4 < len(arr)):
        x1 = arr[i] - (arr[i+2] / 2)
        y1 = arr[i+1] - (arr[i+3] / 2)
        x2 = arr[i] + arr[i+2] / 2
        y2 = arr[i+1] + arr[i+3] / 2
        can.create_rectangle(x1, y1, x2, y2, fill = 'red')
    i += 1
window.mainloop()
```
